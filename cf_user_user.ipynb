# Mean-Centered Overlapping Collaborative Filtering

```python
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

DATA_DIR = Path("data/ml-latest-small/ml-latest-small")
ratings_path = DATA_DIR.joinpath("ratings.csv")
ratings = pd.read_csv(ratings_path)

ratings.drop(columns=["timestamp"], inplace=True)

user_item_matrix = ratings.pivot_table(index="userId", columns="movieId")

user_means = user_item_matrix.mean(axis=1)
mean_centered_matrix = user_item_matrix.sub(user_means, axis=0)

def temp(m):
    n_users = len(m)
    
    sim_m = np.zeros((n_users, n_users))
    
    for i in range(n_users):
        for j in range(n_users):
            common = ~m.iloc[i].isna() & ~m.iloc[j].isna()
        
        if common.any():
            rating_i = m.iloc[i][common]
            rating_j = m.iloc[j][common]
            
            rating_i = np.array(rating_i)
            rating_j = np.array(rating_j)
            
            sim = cosine_similarity([rating_i], [rating_j])[0][0]
            print(sim)
            sim_m[i, j] = sim
            sim_m[j, i] = sim
    return sim_m

sim = temp(mean_centered_matrix)
```
